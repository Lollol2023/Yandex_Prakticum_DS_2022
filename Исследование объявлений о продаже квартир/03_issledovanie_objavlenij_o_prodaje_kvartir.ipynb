## Предобработка данных

```python
#Переведем значения в столбце 'last_price' в миллионы рублей, сохраним как новый столбец 
#и изменим тип данных на целочисленный в столбце 'last_price':
df['last_price_mln_rub'] = (df['last_price']/1000000).round(3)
pd.to_numeric(df['last_price'],downcast = 'integer')
display(df[['last_price_mln_rub', 'last_price']].head(8))
```

```python
#Заполним пропуски в столбце 'is_apartment' значением False. После этого тип переменной сменится на 'bool'.
df['is_apartment'].fillna(value = False, inplace = True)
```

```python
#Выясним медианное значение параметра высоты потолков по всему датафрейму
display('Медианная высота потолков:', df['ceiling_height'].median().round(1))
```

```python
#Заполним пропуски в столбце 'ceiling_height' медианным значением
df['ceiling_height'].fillna(value = df['ceiling_height'].median(), inplace = True)
```

```python
#Заполним пропуски в столбце 'balcony' нулями Потом изменим тип переменной на целочисленную
df['balcony'].fillna(value = 0, inplace = True)
pd.to_numeric(df['balcony'],downcast = 'integer')
```

```python
#В столбце 'floors_total' 86 строк содержат пропуски. Это не очень существенная часть датафрейма. 
#Можно удалить эти строки или заменить на соответствующие значения из столбца 'floor'.
#Потом изменим тип переменной на целочисленную.
df['floors_total'].fillna(value = df['floor'], inplace = True)
pd.to_numeric(df['floors_total'],downcast = 'integer')
```

```python
#Построим сводную таблицу, чтобы посмотреть распределение числа строк и средних значений 
#площади кухни и жилой площади, 
#в зависимости от количества комнат и принадлежности к категории студии
df_pivot1 = df.pivot_table(index = ['studio', 'rooms'], values = ['kitchen_area','living_area'], aggfunc = ['count', 'mean'])
display(df_pivot1)
```

Мы видим, что квартиры, где число комнат == 0, могут как быть, так и не быть студиями. 

Однако в любом случае средняя площадь кухни там ==0. 

Также видим, что во всех студиях (как с 0, так и с 1 комнатой) средняя площадь кухни тоже == 0.
    
Также видим, что NaN !== 0 - пропущенные/незаполненные ячейки встречаются в сегменте студий, что видно из таблицы.

Видим также, как увеличивается средняя площадь кухни с ростом числа комнат.

Продолжим заполнять пропущенные значения.

```python
#В столбце 'kitchen_area' заменим пропущенные значения на 0 для строк с параметром 'studio == True', 
#чтобы охватить студии с указанным количеством комнат == 1
df.loc[(df['studio'] == True) , 'kitchen_area'] = 0
```

```python
#вообще не очень логично, что в квартире значится 0 комнат. 
#Даже студия с открытой планировкой подразумевает одно, но обособленное помещение
#Заменим 0 в столбце 'rooms' на 1
df.loc[(df['rooms'] == 0), 'rooms'] = 1
```

```python
#Подставим соответствующие количеству комнат средние значения в ячейки с пропусками столбца 'kitchen_area'.
df.loc[(df['rooms'] == 1) & ~ (df['kitchen_area'] > 0), 'kitchen_area'] = df.loc[df['rooms'] ==1, 'kitchen_area'].mean().round(1)
df.loc[(df['rooms'] == 2) & ~ (df['kitchen_area'] > 0), 'kitchen_area'] = df.loc[df['rooms'] ==2, 'kitchen_area'].mean().round(1)
df.loc[(df['rooms'] == 3) & ~ (df['kitchen_area'] > 0), 'kitchen_area'] = df.loc[df['rooms'] ==3, 'kitchen_area'].mean().round(1)
df.loc[(df['rooms'] == 4) & ~ (df['kitchen_area'] > 0), 'kitchen_area'] = df.loc[df['rooms'] ==4, 'kitchen_area'].mean().round(1)
df.loc[(df['rooms'] == 5) & ~ (df['kitchen_area'] > 0), 'kitchen_area'] = df.loc[df['rooms'] ==5, 'kitchen_area'].mean().round(1)
df.loc[(df['rooms'] == 6) & ~ (df['kitchen_area'] > 0), 'kitchen_area'] = df.loc[df['rooms'] ==6, 'kitchen_area'].mean().round(1)
df.loc[(df['rooms'] == 7) & ~ (df['kitchen_area'] > 0), 'kitchen_area'] = df.loc[df['rooms'] ==7, 'kitchen_area'].mean().round(1)
df.loc[(df['rooms'] == 8) & ~ (df['kitchen_area'] > 0), 'kitchen_area'] = df.loc[df['rooms'] ==8, 'kitchen_area'].mean().round(1)
df.loc[(df['rooms'] == 9) & ~ (df['kitchen_area'] > 0), 'kitchen_area'] = df.loc[df['rooms'] ==9, 'kitchen_area'].mean().round(1)
```

```python
#Заполним пропуски в столбце 'living_area', рассчитав среднее соотношение общей площади к жилой площади.
mean_living_total = df['living_area'].mean()/df['total_area'].mean() #средний коэффициент по всему датафрейму
display('Среднее соотношение общей и жилой площади:', mean_living_total) #равен 0,6

median_living_total = df['living_area'].median()/df['total_area'].median() #средний коэффициент по всему датафрейму
display('Медианное соотношение общей и жилой площади:', median_living_total) #равен 0,6

df['living_area'].fillna(value = df['total_area']* mean_living_total, inplace = True)
```

```python
#В столбце 'days_exposition' изменим тип переменной на целочисленную.
pd.to_numeric(df['days_exposition'],downcast = 'integer')
```

```python
#Изменим формат данных в столбце с датой
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format = '%Y-%m-%d')
```

```python
#Пропуски в столбце 'locality_name' решено просто удалить - их всего 50. 
#Исправим двойной вариант написания названия нас.пункта методом .replace()
df = df.dropna(subset = ['locality_name']).reset_index(drop=True)
df['locality_name'] = df['locality_name'].replace('посёлок', 'поселок', regex=True)
```

## Расчёты и добавление результатов в таблицу

Добавим новые столбцы в датафрейм: 

* cтоимость кв.м, 
* соотношение "жилая/общая площадь", 
* соотношение "площадь_кухни/общая_площадь".
